```python
import os
from langchain.llms import OpenAI
from langchain.agents import AgentType, initialize_agent
from langchain.tools import Tool
from langchain.memory import ConversationBufferMemory
from dotenv import load_dotenv

load_dotenv()
OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
```

```python
pip install langchain_community
```

```python
llm = OpenAI(api_key=OPENAI_API_KEY, temperature=0.7)
```

```python
from langchain_core.memory import ConversationBufferMemory

memory = ConversationBufferMemory(return_messages=True)
```

```python
from langchain.tools import tool
from langchain.utilities import SerpAPIWrapper

search_tool = Tool(
    name="Google Search",
    func=SerpAPIWrapper(api_key="YOUR_SERPAPI_KEY").run,
    description="Searches Google for real-time information."
)
```

```python
tools = [search_tool]

agent = initialize_agent(
    tools=tools,
    llm=llm,
    agent=AgentType.ZERO_SHOT_REACT_DESCRIPTION,
    memory=memory,
    verbose=True
)

response = agent.run("What is the latest stock price of Tesla?")
print(response)
```

```python
from langchain.chains import SequentialChain
from langchain.prompts import PromptTemplate

# Step 1: Search for the latest news on Tesla
news_prompt = PromptTemplate(
    input_variables=["company"],
    template="Find the latest financial news about {company}."
)

# Step 2: Summarize the findings
summary_prompt = PromptTemplate(
    input_variables=["news"],
    template="Summarize the following news: {news}"
)

search_chain = SequentialChain(
    chains=[news_prompt, summary_prompt],
    input_variables=["company"],
    verbose=True
)

response = search_chain.run("Tesla")
print(response)
```
